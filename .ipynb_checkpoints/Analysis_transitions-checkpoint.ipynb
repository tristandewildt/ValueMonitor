```python
best_number_of_topics = 200
number_of_documents_in_analysis = 2000

number_of_words_per_topic_to_show = 20
number_of_words_per_topic = 20

focus_on_articles_containing_keywords = True
requilibrate_articles_containing_keywords = True

dict_anchor_words = {
"Sustainability" : ["sustainability", "sustainable", "renewable", "renewable energy", 'durability', 'durable'],        
"Economic viability" : ["economic viability", "economic", "economic potential", "costs", "cost effective", "cost"],
"Affordability" : ["affordability", "affordable", "energy security", "low cost", "income", "poor", 
                   "poverty", "low income", "accessibility"],
"Availability" : ["availability", "reliability", "reliable", "security of supply"],
"Safety & Health" : ["safety", "health", "accident", "accidents"],
"Justice" : ["justice", "fairness", "social equity", "injustice", "injustices",
            "social fairness", "inequality", "inequalities",],
"Privacy & security" : ["privacy", "privacy concerns", "privacy preserving", "data privacy", 
             "privacy perservation", "cyber"],

}

list_rejected_words = ["energy initiative", "international energy", "emissions", "fuel", "gas", "greenhouse", 
                       "of energy", "fuels", "fossil", "greenhouse gas", "households", "household", "households in", 
                       "households and", "drpt", "2008 drpt", "water availability", "argue", "argue that", "we argue",
                       "ieee", "2017 ieee", "things", "cloud", "the internet", "real time", "of smart",
                       "rural", "africa", "in rural", "sub saharan", "saharan", "of rural", "livelihoods", "women", "saharan africa",
                       "neural", "neural network", "neural networks", "validity", "and system",
                       "john wiley", "wiley", "sons ltd", "sons", "erp environment", "erp", "john", "wiley sons", "and erp", "ltd and",
                       "no abstract", "available nan", "massachusetts institute", "technology nan", "massachusetts",
                      ]

'''--------------------------------------------------------------------------''' 
df_reduced = df

if focus_on_articles_containing_keywords == True:
    list_words = list(dict_anchor_words.values())
    pattern = '|'.join([item for sublist in list_words for item in sublist])
    df_reduced = df_reduced[df_reduced['[Text_for_analysis]'].str.contains(pattern)]
        
if requilibrate_articles_containing_keywords == True:
    dict_frequencies = {}
    for key, value in dict_anchor_words.items():
        pattern = '|'.join(value)
        dict_frequencies[key] = len(df_reduced[df_reduced['[Text_for_analysis]'].str.contains(pattern)])
    dict_frequencies = dict(sorted(dict_frequencies.items(), key=lambda item: item[1]))


    df_focused = pd.DataFrame(columns=df_reduced.columns.tolist())
    number_of_articles_still_to_take = number_of_documents_in_analysis
    average_number_articles_to_take = round(number_of_documents_in_analysis / len(dict_anchor_words))
    for key, value in dict_frequencies.items():
        pattern = '|'.join(dict_anchor_words[key])
        df_this_value = df_reduced[df_reduced['[Text_for_analysis]'].str.contains(pattern)]
        if len(df_this_value) > average_number_articles_to_take:
            df_this_value = df_this_value.sample(n=average_number_articles_to_take)
        else:
            remaining_articles_not_taken = average_number_articles_to_take - len(df_this_value)
            average_number_articles_to_take = average_number_articles_to_take + round(remaining_articles_not_taken / len(dict_anchor_words))
        df_focused = df_focused.append(df_this_value)
    df_focused = df_focused.drop_duplicates(subset='[Text_for_analysis]', keep="first")
    number_of_documents_in_analysis = len(df_focused)
    df_reduced = df_focused
print(len(df_reduced))

model_and_vectorized_data = make_anchored_topic_model(df_reduced, best_number_of_topics, min(number_of_documents_in_analysis, len(df_reduced)), dict_anchor_words, list_rejected_words)
topics = report_topics(model_and_vectorized_data[0], dict_anchor_words,number_of_words_per_topic )
df_with_topics = create_df_with_topics(df, model_and_vectorized_data[0], model_and_vectorized_data[1], best_number_of_topics)
```

```python
copy_df_with_topics = df_with_topics.copy()
copy_dict_anchor_words = dict_anchor_words.copy()

df_with_topics_freq = copy_df_with_topics.set_index('[Date]').resample(resampling).size().reset_index(name="count")
df_with_topics_freq = df_with_topics_freq.set_index('[Date]')

df_frequencies = copy_df_with_topics.set_index('[Date]')
df_frequencies = df_frequencies.resample(resampling).sum()

#list_topics = list(range(len(copy_dict_anchor_words)))
#df_frequencies = df_frequencies[list_topics]

df_frequencies = df_frequencies.div(df_with_topics_freq["count"], axis=0)
combined_df = pd.concat([df_frequencies, df_with_topics_freq], axis=1)
combined_df = combined_df.fillna(0)

x = pd.Series(combined_df.index.values)
x = x.dt.to_pydatetime().tolist()

combined_df = combined_df * 100
combined_df = combined_df.drop("count", 1)
```

```python
number_of_words_per_topic = 5

list_values = []
for key, value in dict_anchor_words.items(): 
    list_values.append(key)

list_topics = []
for i, topic_ngrams in enumerate(model_and_vectorized_data[0].get_topics(n_words=number_of_words_per_topic)):
    topic_ngrams = [ngram[0] for ngram in topic_ngrams if ngram[1] > 0]
    
    if len(list_values) > i:
            list_topics.append("Topic #{} ({}): {}".format(i, list_values[i], ", ".join(topic_ngrams)))
    else:
        list_topics.append("Topic #{}: {}".format(i, ", ".join(topic_ngrams)))
```
