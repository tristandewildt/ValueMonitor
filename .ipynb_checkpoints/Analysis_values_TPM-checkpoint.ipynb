```python
import fuzzywuzzy
from fuzzywuzzy import process
import chardet
import numpy as np

''' Then we use FuzzyWuzzy to match inconsistent author data'''

df['[Authors]'] = df['[Authors]'].str.lower()
df['[Authors]'] = df['[Authors]'].str.strip()

# First we need to separate the author data into lists
df_list_authors = df 
for index, row in df.iterrows():
    df_list_authors.at[index,'[Authors]'] = row['[Authors]'].split(", ")
```

```python
#df_list_authors
```

```python
''' Then we try to find matching ones '''
df = df_list_authors

# first we create a list of unique authors
list_authors_by_articles = df['[Authors]'].tolist()
list_authors_by_articles = [item for sublist in list_authors_by_articles for item in sublist]
list_authors_by_articles= list(set(list_authors_by_articles))

# the we try to find matching values
min_ratio = 76
dict_transformed_authors = {}
#list_authors_by_articles = list_authors_by_articles[300:]
for author in list_authors_by_articles:
    matches = fuzzywuzzy.process.extract(author, list_authors_by_articles, limit=10, scorer=fuzzywuzzy.fuzz.token_sort_ratio)
    close_matches = [matches[0] for matches in matches if matches[1] >= min_ratio]
    for i in close_matches:
        dict_transformed_authors[i] = close_matches[0]
    list_authors_by_articles = [element for element in list_authors_by_articles if element not in close_matches]

# Maybe also remove the authors from 'list_authors_by_articles' to make the process quicker
```

```python
''' Then we go through the df and change names where needed'''

df_merged_authors = df 
for index, row in df_merged_authors.iterrows():
    these_authors = row['[Authors]']
    for author in these_authors:
        if author in dict_transformed_authors:
            these_authors[these_authors.index(author)] = dict_transformed_authors[author]
    #these_authors = ", ".join(these_authors)
    df_list_authors.at[index,'[Authors]'] = these_authors
    #.index("bar")
```

```python
''' Here we could add something to specify the type of graph that we want to have, 
    e.g. which value, which type of data, etc. '''

number_of_topics_selected = 0

for i in range(number_of_topics_selected):
    df_with_topics_graph = df_with_topics[df_with_topics[i] == 1]
    #df_value_sdf_with_topics_graphpecific = df_with_topics_graph[['[Text_for_analysis]', '[Date]', '[Authors]']]
    # change name to match VSO viewer format
    #df_value_specific=df_value_specific.rename(columns = {'[Authors]':'Authors'})
    #df['Authors'] = df['[Authors]']
    #df_value_specific.to_csv('data/Value_'+str(i)+'_'+str(file_name))
```

```python
# Create graph
g = Graph(directed=False)

# Add x vertices
list_authors_for_graph = []
for index, row in df_with_topics_graph.iterrows():
    these_authors = row['[Authors]']
    for author in these_authors:
        list_authors_for_graph.append(author)
        
list_authors_for_graph = list(set(list_authors_for_graph))
dict_authors_for_graph = {}
counter = 0
for author in list_authors_for_graph:
    dict_authors_for_graph[author]=counter
    counter += 1

g.add_vertices(len(list_authors_for_graph))

# Add ids and labels to vertices
for i in range(len(g.vs)):
    g.vs[i]["id"]= i
    g.vs[i]["label"]= list_authors_for_graph[i]
```

```python
list_edges = []
for index, row in df_with_topics_graph.iterrows():
    start_counter = 1
    these_authors = row['[Authors]']
    for first_author in these_authors:
        counter = start_counter
        while counter < len(these_authors):
            second_author = these_authors[counter]
            new_tuple = (dict_authors_for_graph[first_author], dict_authors_for_graph[second_author])
            list_edges.append(new_tuple)
            counter += 1
        start_counter += 1
```

```python
# Then we create list of weights. This is the frequency of counts of edges.

import collections

dict_weights = collections.Counter(list_edges)

list_edges_unique = []
weights = []
for key, value in dict_weights.items():
    weights.append(value)
    list_edges_unique.append(key)
    
# Add edges
g.add_edges(list_edges_unique)

# Add weights and edge labels
g.es['weight'] = weights
#g.es['label'] = weights
```

```python
visual_style = {}

out_name = "graph.png"

# Set bbox and margin
visual_style["bbox"] = (400,400)
visual_style["margin"] = 27

# Set vertex colours
visual_style["vertex_color"] = 'white'

# Set vertex size
visual_style["vertex_size"] = 3

# Set vertex lable size
visual_style["vertex_label_size"] = 8

# Don't curve the edges
visual_style["edge_curved"] = True

# Set the layout
my_layout = g.layout_lgl()
visual_style["layout"] = my_layout

# Plot the graph
plot(g, out_name, **visual_style)
```
